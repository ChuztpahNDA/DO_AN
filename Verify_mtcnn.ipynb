```python
def one_hot_encode(dataframe:list):
    dict_emotion = {'angry':1, 'anger':1, 'disgust':2, 
                    'fear':3, 'happy':4, 
                    'sadness':5, 'sad': 5, 
                    'surprise':6, 'neutral':0}
    result = []
    for i in range(len(dataframe)):
        if type(dataframe[i])  != int:
            result.append(dict_emotion[dataframe[i]])
    return result
```
