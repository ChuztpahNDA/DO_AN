```python
from deepface import DeepFace
from PIL import Image
import matplotlib.pyplot as plt
import gdown
```

```python
image_path = 'user/Aaron_Eckhart_0001.jpg'
db_path = 'user/db'
```

### Face recognition: model = FaceNet, backend = MTCNN

```python
# image_path đường dẫn tới hình ảnh cần tìm trong database
# db_path đường dẫn tới database
%time
df_consine = DeepFace.find(image_path, db_path, model_name='Facenet') 
df_euclidean = DeepFace.find(image_path, db_path, distance_metric='euclidean', model_name='Facenet')
df_euclidean_l2 = DeepFace.find(image_path, db_path, distance_metric='euclidean_l2', model_name='Facenet')
```

```python
df_consine
```

```python
df_euclidean
```

```python
df_euclidean_l2
```

### euclidean_l2_FaceNet

```python
distence = df_euclidean_l2.drop(['identity'], axis=1)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.imshow(Image.open(df_euclidean_l2.identity[0]))
plt.title(f'distance: {distence.values[0].tolist()[0]}')
plt.subplot(1,2,2)
plt.imshow(Image.open(image_path))
plt.title('original')
```
