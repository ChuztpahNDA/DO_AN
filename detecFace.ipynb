```python
from deepface import DeepFace
import matplotlib.pyplot as plt 
from glob import glob
```

```python
img_paths = glob('./user/detecFace/*')
```

```python
?DeepFace.detectFace
```

```python
def detecface_(back_end : str, paths:list):
    faces = []
    for path in paths:
        try:
            faces.append(DeepFace.detectFace(path, detector_backend = back_end))
        except:
            faces.append(-1)
    return faces
        
```

```python
face_mtcnn = detecface_(back_end = 'mtcnn', paths = img_paths[0:20])
```

```python
len(face_mtcnn)
```

```python
f = plt.figure()
count = 0
for i in range(len(face_mtcnn)):
    if type(face_mtcnn[i]) != int:
        f.add_subplot(1,1,1)
        plt.imshow(face_mtcnn[i])
        plt.show(block=True)
        print('---------------------------------')
    else: count += 1
print(count)
```
