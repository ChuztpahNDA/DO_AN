```python
def detecface_(back_end : str, paths:list):
    faces = []
    for path in paths:
        try:
            faces.append(DeepFace.detectFace(path, detector_backend = back_end))
        except:
            faces.append(-1)
    return faces
        
```

```python
face_mtcnn = detecface_(back_end = 'mtcnn', paths = train_path[0:20])
```
